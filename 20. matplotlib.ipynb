### `Write a list comprehension such that even number should be replaced by 0 and keep the odd numbers as it is.`

### `Write a program to reverse the string elements in the list`
- input: ['abc', 'def', 'ghi']
- output: ['cba', fed', 'igh']

````python
input = ['abc', 'def', 'ghi']
output = []
for ele in input:
  rev = ''
  for i in ele:
    rev = i + rev
  output.append(rev)
output

#inp=['Utsarga','Mandish','Waibhawa']
#print([i[::-1] for i in inp])
````
